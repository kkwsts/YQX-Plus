```python
# find outliers in the targets distribution using statistical methods
import numpy as np

targets_distribution = np.load("artifacts/targets_distribution.npy")
# Calculate mean and standard deviation for each target dimension
means = np.mean(targets_distribution, axis=0)
stds = np.std(targets_distribution, axis=0)

# Define outlier threshold as 3 standard deviations from mean
threshold = 3

# Find outliers for each target dimension
outliers = []
for i in range(4):
    target_data = targets_distribution[:, i]
    z_scores = np.abs((target_data - means[i]) / stds[i])
    outlier_indices = np.where(z_scores > threshold)[0]
    outliers.append((i, outlier_indices))

# Print summary of outliers found
print("\nOutlier Analysis (using 3 standard deviations):")
print("-" * 50)
for target_idx, indices in outliers:
    print(f"\nTarget {target_idx}:")
    print(f"Number of outliers: {len(indices)} out of {len(targets_distribution)}")
    if len(indices) > 0:
        print(f"Outlier values: {targets_distribution[indices, target_idx]}")
        print(f"Mean: {means[target_idx]:.3f}")
        print(f"Std Dev: {stds[target_idx]:.3f}")


```
